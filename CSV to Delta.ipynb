```python
import os

print(
  os.listdir(
    '/Volumes/kroll_catalog/input_schema/kroll_input_volume/'
  )
)
```

```python
all_files = os.listdir(
    '/Volumes/kroll_catalog/input_schema/kroll_input_volume/'
  )

for i in all_files:
    file_name = i.split(".")[0]
    print(file_name)
    table_name = f"kroll_catalog.input_schema.{file_name}"

    full_path = f"/Volumes/kroll_catalog/input_schema/kroll_input_volume/{i}"

    spark.sql(f"""CREATE TABLE if NOT EXISTS {table_name};""")

    spark.sql(f"""COPY INTO {table_name}
    FROM '{full_path}'
    FILEFORMAT = CSV
    FORMAT_OPTIONS ('header' = 'true', 'inferSchema' = 'true')
    COPY_OPTIONS ('mergeSchema' = 'true'); """)

```
